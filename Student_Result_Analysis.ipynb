```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Student_Scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()   #tells datatype of each column
```

```python
df.isnull().sum() #returns count of null values in each column
```

# Drop unnamed column

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

# Change weekly study hours column

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
# from the above chart we have analysed that the number of females in the data is more than the number of males 
```

# Parent education impact on scores of students

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Relation between parent's education and student's score")
plt.show()
```

```python
# the above plot shows that education of parents have a good impact on their child's score,
# as it can be clearly seen that children whose parents have higher educational qualifications are scoring more on an average
# children whose parents have done master's degree have higher average marks, 
# whereas childern whose parents just have studied till high school have lower average marks
```

# Impact of parent martial status on child education

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Relation between parent's marital status and student's score")
plt.show()
```

```python
# from above chart we can conclude that there is negligible impact on student's score due to their parents martial status
```

# Ethnic group percentage distribution

```python
df["EthnicGroup"].unique()
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
# print(groupA["EthnicGroup"])

l=["group A","group B","group C","group D","group E"] #for creating labels
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%") #show labels along with value upto 2 decimal places
plt.title("Distribution of Ethnic groups")
plt.show()
```
